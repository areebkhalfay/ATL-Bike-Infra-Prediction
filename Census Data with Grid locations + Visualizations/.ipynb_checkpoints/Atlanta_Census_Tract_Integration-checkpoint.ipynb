```python
import pandas as pd
import geopandas as gpd
```

```python
csv_path = "NYC_GEOINFO2023.GEOINFO-Data.csv"
df = pd.read_csv(csv_path).drop(index=0)
```

```python
df.head()
```

```python
gdf = gpd.read_file("tl_2023_36_tract.shp")
```

```python
gdf.head()
```

```python
df['GEOID'] = df['GEO_ID'].astype(str)
gdf['GEOID'] = gdf['GEOIDFQ'].astype(str)
```

```python
df = df.drop('GEO_ID', axis=1)
gdf = gdf.drop('GEOIDFQ', axis=1)
```

```python
merged_gdf = gdf.merge(df, on='GEOID')
nyc_geo_census_polygons = merged_gdf[['GEOID', 'NAMELSAD', 'AREALAND', 'AREAWATR', 'AREALAND_SQMI', 'AREAWATR_SQMI', 'geometry']]
```

```python
nyc_geo_census_polygons.to_csv("ny_census_tracts_with_data.csv", index=False)
```
